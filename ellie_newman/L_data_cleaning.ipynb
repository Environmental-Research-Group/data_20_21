```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json("https://data.cityofchicago.org/resource/5neh-572f.json?$limit=201000&$offset=909500")
station_loc = pd.read_json("https://data.cityofchicago.org/resource/8pix-ypme.json")
#https://data.cityofchicago.org/Transportation/CTA-System-Information-List-of-L-Stops/8pix-ypme
```

```python
df['date'] = pd.to_datetime(df['date'],format='%y/%m/%d %I:%M%p')
df['month'] = df.date.dt.month
df['year'] = df.date.dt.year
df['day'] = df.date.dt.day
```

```python
group_mon = df.groupby(by=[df.month, df.year])
group_date = df.groupby(['date'])
```

```python
mean_rides_per_mon = df.groupby(by=[df.month, df.year]).mean()
%matplotlib
mean_rides_per_mon.plot.line(x="date",y="rides")
```

```python
reset_index = mean_rides_per_mon.reset_index()
```
