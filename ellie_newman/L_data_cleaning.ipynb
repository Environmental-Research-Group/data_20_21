```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json("https://data.cityofchicago.org/resource/5neh-572f.json?$limit=200000&$offset=909000")
```

```python
df.describe()
```

```python
df.head()
```

```python
#check for NaN
df["date"].isnull().sum()
df["rides"].isnull().sum()
df["stationname"].isnull().sum()
df["station_id"].isnull().sum()
```

```python
df[df["rides"]==0]
```

```python
len(df[df["rides"]==0])
```

```python
one_day=df[df["date"]=="2020-01-20"]
```

```python
Ohare=df[df["stationname"]=="O'Hare Airport"]
```

```python
%matplotlib
Ohare.plot.line(x="date",y="rides")
#O'Hare station closed in Oct 2019
```

```python
%matplotlib
plt.barh(one_day["stationname"],one_day["rides"])
#O'Hare station closed in Oct 2019
```
